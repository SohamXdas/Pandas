## Filterring data 

```python
## Now for example we only want to see pokemon who are grass type the syntax for that is as follows
df.loc[df['Type 1'] == 'Grass']
```

```python
## We can also use multiple filters by using other logical arguments like (& | or) etc that will look something like the following
```

```python
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison')]
```

##### BTW "and" can be written as (and or &)
##### Similarly "or" can be written as (OR / |)
##### Only applicable in pandas library

#### Now let's see how to filter only the Pokemon who has mega in there name 

```python
df.loc[df['Name'].str.contains('Mega')]
```

```python
## And now for example we need the names of the pokemon excluding the mega ones we do the following
df.loc[~df['Name'].str.contains('Mega')]
```

#### Now let's look at some other complex data filtration problems
##### What if we want to see all the Pokemon who is either grass-type or fire type 
##### We use the re pacakge (regular expression) pacakge 

```python
import re
df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)]
## the flags=re.I is a flag which is used to ignore the case sensitivity in a data
```

##### Now let's filter the Pokemon which's name starts with "Pi"

```python
df.loc[df['Name'].str.contains('^pi[a-z]', flags=re.I, regex=True)]
## The ^ Stands for the start of the word also called an anchor
```

## Conditional Changes

##### For example we want to change the name of type 2 water to Squirt

```python
df.loc[df['Type 1'] == 'Grass', 'Type 1'] = 'Squirt'
df
```

```python
## Now lets turn it back to normal
df.loc[df['Type 1'] == 'Squirt', 'Type 1'] = 'Grass'
df
```

## Agrigate Statistics(Groupby)

#### This is used to assess standard statistical data.  
##### For example if we want to see which type of Pokemon has the highest defence we can do the following we can take the highest mean

```python
df = pd.read_csv('modified.csv')
df.groupby(['Type 1']).mean(numeric_only=True).sort_values('Defense', ascending=False)
```
